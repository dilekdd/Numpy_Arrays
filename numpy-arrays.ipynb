Multiply two lists without using numpy arrays

```python
import numpy as np
a = [4, 3, 2, 1]
b = [2, 3, 4, 5]

ab = []

for i in range(0, (len(a))):
    ab.append(a[i] * b[i])
print(ab)
```

multiply two numpy arrays

```python
import numpy as np
a = np.array([4, 3, 2, 1])
b = np.array([2, 3, 4, 5])
print(a*b)
```
